```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import r2_score, mean_squared_error
from mpl_toolkits.mplot3d import Axes3D

parasitos = pd.read_csv("parasito.csv")
parasitos.head()
```

```python
dataX2 = pd.DataFrame()
dataX2['Temperatura'] = parasitos['Temperatura']
dataX2['Humedad'] = parasitos['Humedad']
XY_train = np.array(dataX2)
Z_train = parasitos['Recuento']
regr = linear_model.LinearRegression()
regr.fit(XY_train,Z_train)
predict_z = regr.predict(XY_train)
regr.score(XY_train,Z_train)
```

```python
fig = plt.figure()
ax = Axes3D(fig)

ax.scatter(XY_train[:, 0], XY_train[:, 1], Z_train, c='blue')

ax.scatter(XY_train[:, 0], XY_train[:, 1], predict_z, c='red')

ax.view_init(elev=40, azim=-30)
```
