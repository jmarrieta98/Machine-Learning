## Operaciones con dos dataframes

Generamos dos dataframes: 

```python
exam_data  = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
        'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
        'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

# Creación combinando un diccionario y un array: 
df1 = pd.DataFrame(exam_data , index=labels)

# Creación directa desde un diccionario:  
df2 =pd.DataFrame({'name': ['Dima', 'Katherine', 'Laura', 'Kevin', 'Jonas', 'Francisco', 'Gabriella'],
                    'country':['India', 'USA', 'Spain', np.nan, 'Israel', 'Spain', 'Portugal']})
```

```python
df1
```

```python
df2
```

Combina los dataframes df1 y df2 manteniendo **SOLO** las filas de los que estén en los dos dataframes

```python
pd.merge(df1,df2,how='inner',on='name', indicator=True)
```

Combina los dataframes df1 y df2 manteniendo **TODAS** las filas de los que estén en los dos dataframes

```python
pd.merge(df1,df2,how='outer',on='name', indicator=True)
```

Ahora solo las filas combinadas de los que esten en el primer dataframe especificado

```python
dfmerge = pd.merge(df1,df2, on='name',how='outer', indicator=True)
```

```python
dfmerge[dfmerge['_merge']=='left_only']
```

Ahora solo las filas de los especificados en el segundo dataframe

```python
dfmerge[dfmerge['_merge']=='right_only']
```

Relaciona los dos dataframes mediante un left join

```python
pd.merge(df1,df2,on='name', how='left', indicator=True)
```

Concatena ambos dataframes

```python
pd.concat([df1,df2])
```
