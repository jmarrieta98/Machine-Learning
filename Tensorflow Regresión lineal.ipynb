```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
W = tf.Variable(12.0)
b = tf.Variable(-6.1)
```

```python
weight = 2.5
bias = 1.0
```

```python
data = 100
inputs  = tf.random.normal(shape=[data])
noise   = tf.random.normal(shape=[data])
outputs = inputs * weight + bias + noise
```

```python
Ws, bs = [], []

epochs = range(30)

learning_rate=0.1

for epoch in epochs:
  with tf.GradientTape() as tape:
    loss = tf.reduce_mean(tf.square(outputs-(W*inputs+b)))
  dW, db = tape.gradient(loss, [W,b])
  Ws.append(W.numpy())
  bs.append(b.numpy())

  W.assign_sub(learning_rate * dW)
  b.assign_sub(learning_rate * db)

  print("=> epoch %2d: w_true= %.2f, w_pred= %.2f; b_true= %.2f, b_pred= %.2f, loss= %.2f" % (epoch+1, weight, W.numpy(), bias, b.numpy(), loss.numpy()))
  if (epoch) % 2 == 0:
    plt.scatter(inputs, outputs, c='b')
    #plt.scatter(inputs, model(inputs), c='m') 
    plt.plot(inputs, (W*inputs+b),  c='r', label ='Fitted line')
    
```
