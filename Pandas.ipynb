```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import linear_model
from sklearn.metrics import r2_score
alumnos = pd.read_csv("/content/alumnos.csv")

alumnos.head()
```

```python
regr = linear_model.LinearRegression()
regr.fit( np.array(alumnos["Peso"]).reshape(-1,1),alumnos["Altura"])
y_predict = regr.predict(np.array(alumnos["Peso"]).reshape(-1,1))
r2_score(alumnos["Altura"],y_predict)
```
